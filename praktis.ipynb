```python
import librosa
import matplotlib.pyplot as plt
import numpy as np

audio, sr = librosa.load('B2.wav', sr=22050)
# librosa.display.waveshow(audio)

#time-domain to frequency-domain
fft = np.fft.fft(audio)
magnitude = np.abs(fft)
frequency = np.linspace(0, sr, len(magnitude))

left_frequency = frequency[:int(len(frequency)/2)]
left_magnitude = magnitude[:int(len(magnitude)/2)]
print(np.max(left_frequency))

# plt.plot(left_frequency, left_magnitude)
# plt.xlabel('Frequency')
# plt.ylabel('Magnitude')
# plt.xlim(0, left_frequency[-1])
# plt.show()

# convert to stft spectrogram
n_fft = 2048
hop_length = 512

# stft = librosa.core.stft(audio, n_fft=n_fft, hop_length=hop_length)
# spectrogram = np.abs(stft)
# log_spectrogram = librosa.amplitude_to_db(spectrogram)
# librosa.display.specshow(log_spectrogram, n_fft=n_fft, hop_length=hop_length)
# plt.xlabel('Time')
# plt.ylabel('Frequency')
# plt.colorbar()
# plt.show()
```

```python
#to MFCC
mfcc = librosa.feature.mfcc(y=audio, n_fft=n_fft, hop_length=hop_length, n_mfcc=13)
librosa.display.specshow(mfcc, sr=sr, hop_length=hop_length)
plt.xlabel('Time')
plt.ylabel('MFCC')
plt.colorbar()
plt.show()
```
